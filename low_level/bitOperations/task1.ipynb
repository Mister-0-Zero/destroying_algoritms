```python
def decimalToBin(number):
    res = ""
    while number > 0:
        res = str(number % 2) + res
        number //= 2
    return res

def fifeBitNumber(number):
    binary = decimalToBin(number)
    len_b = len(binary)
    res = (5 - len_b) * "0" + binary
    return res
```
